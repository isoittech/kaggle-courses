# 3.4 幾何分布（Geometric Distribution）



このノートブックでは、統計学と確率論の基礎となる離散確率分布の一つである幾何分布について学ぶ。幾何分布は「初めて成功するまでの試行回数」を表す確率分布だ。

## 3.4.1 幾何分布の基本概念と定義（やさしい解説）

幾何分布（きかぶんぷ）とは、「初めて成功するときまでに、合計で何回チャレンジが必要か」ということを表す確率分布である。

例えば、こんなイメージである。

- コインを投げて、初めて表が出るまで何回投げる？
- くじ引きで、当たりを引くまでに何回引く？

こうした「どれだけ失敗が続くか、そしてようやく一度成功する」…これが幾何分布の世界である。

---

### 幾何分布の特徴

初心者がつまずきやすい言葉を、簡単に説明する。

1. **独立試行の繰り返し**
   
   - 「独立」とは、前の結果が後の結果に全く影響しないこと。
   - 例：「1回めにサイコロで6が出なくても、2回目に6が出る確率は変わらない。」
   - 言い換え：やり直すたびに「新しいスタート」になるイメージ。

2. **二値の結果**（ふたつの結果）
   
   - 毎回のチャレンジで「成功」か「失敗」しかない世界を考える。
   - 例：コインなら「表（成功）」か「裏（失敗）」

3. **成功確率は一定**
   
   - 何度チャレンジしても、成功する確率（p）は毎回おなじ。
   - 例：コインなら何回でも1/2、サイコロで6なら1/6。

4. **初めての成功までを数える**
   
   - 初めて「うまくいった」タイミングが"何回め"なのかを重視。
   - 「1回めで成功する」「3回失敗して4回めで成功」など。

---

### 幾何分布って、どんな場面で使える？

私たちの日常にもたくさん“幾何分布”が隠れている。

- **釣り**：魚が釣れるまでに、何度エサを投げる？
- **営業**：契約がとれるまでに、何人に電話する？
- **くじ引き**：当たりが出るまで、何回くじを引く？
- **サイコロ**：6が出るまで、何回サイコロをふる？

どれも「何回“やってみて”、初めてうまくいく？」という質問である。

---

### 幾何分布のパラメータ（どうやって特徴づける？）

幾何分布は、とてもシンプル。必要な数字はたったひとつである。

- **p**：1回のチャレンジで成功する確率（0 ＜ p ≦ 1）

たとえば…

- 公平なコインで表が出る確率：p = 0.5
- サイコロで6が出る確率：p = 1/6
- 10％の確率で当たるくじ：p = 0.1

この“p”が決まれば、幾何分布は完全に決まる。

---

### 用語のやさしい解説（はじめての人向け）

- **確率分布（かくりつぶんぷ）**：たとえば「サイコロをふったとき、1が出るか2が出るか、それぞれどれぐらいの確率か？」という“おおきな全体のルール”のようなもの。
- **試行（しこう）**：「やってみること」。サイコロをふる1回、コインをなげる1回。
- **成功確率（せいこうかくりつ）**：“あたり”を引く確率。1回でうまくいくチャンス。

---

### 実世界での役立ち方

- たとえば営業職なら、「あと何件電話すれば最初の契約がとれそう？」を予想できる。
- サイコロゲームやくじなど、偶然がからむ状況の「平均〇回で初めて当たる」を考えられる。
- 工場の品質チェックや病院での検査にも「何個検査すれば最初のエラー（病気）が見つかる？」という見通しを立てられる。

このように、**「うまくいくまでに何回がんばればいい？」と考えるとき、幾何分布が役に立つ。**

## 3.4.2 確率質量関数（PMF）（やさしい解説）

### 確率質量関数（PMF）って何だろう？

「確率質量関数（かくりつしつりょうかんすう）」とは、簡単に言うと「ある特定の数値が出る確率を教えてくれる計算式」である。

- ポイント：「この場合は何％くらいのチャンスある？」を知るためのもの。
- ちなみに、英語ではProbability Mass Function（PMF）と呼ばれる。

#### たとえば・・・
- サイコロを振って「3が出る確率は？」といった質問も、PMFが答えられる。

---

### 幾何分布のPMF

幾何分布の場合、「ちょうど k 回目に初めて成功する確率」がPMFで求められる。

幾何分布のPMFの数式は次のようになっている：

$$
P(X = k) = (1-p)^{k-1} \cdot p, \quad k = 1, 2, 3, \ldots
$$

- $X$：はじめて成功するまでの試行回数（「何回目で初めて当たる？」）
- $p$：1回ごとの成功確率（「1回チャレンジしてうまくいく割合」）
- $k$：何回目でチャレンジが成功したか

---

### この式はどんな意味？

この数式の中身を、やさしい言葉で説明する。

- $(1-p)^{k-1}$：最初の$(k-1)$回は全部「失敗」する確率。たとえば3回目で初めて成功なら、2回は連続で失敗しないといけない。
- $p$：$k$回目（つまり最後の回）で「成功」する確率。
- $(1-p)^{k-1} \cdot p$：続けて$k-1$回失敗→その次で当たり！…というストーリー全体の確率。

#### 言い換えイメージ
イメージで言うと「がまんして何度も外れ、ついに当たりを引くまで」の流れをそのまま数式にした感じである。

または...

当たるまでに $k-1$ 回外す確率。

---

### 計算例で実感してみよう

たとえば、コイン投げで表が出る確率が $p = 0.5$ のとき。

- 1回目で表が出る：$P(X=1) = (1-0.5)^{0} \times 0.5 = 1 \times 0.5 = 0.5$
- 2回目ではじめて表：$P(X=2) = (1-0.5)^{1} \times 0.5 = 0.5 \times 0.5 = 0.25$
- 3回目で表：$P(X=3) = (1-0.5)^{2} \times 0.5 = 0.25 \times 0.5 = 0.125$

つまり、失敗が続くほど、だんだん確率が小さくなっていく。

---

### 累積分布関数（CDF）って何？

**累積分布関数（るいせきぶんぷかんすう、CDF）** は、「何回目までに成功してくれる確率はどれくらい？」と質問するときに使う。

幾何分布の場合、この式になる：

$$
P(X \leq k) = 1 - (1-p)^k, \quad k = 1, 2, 3, \ldots
$$

これはつまり「$k$回連続で全部失敗する確率（= $(1-p)^k$）」を1（ぜんぶの確率）から引くやり方。

#### 例
- 3回以内に当たる確率：「1回目も2回目も3回目も外す確率」を引き算すれば出てくる。

---

### 用語のやさしい解説

- **確率質量関数（PMF）**：それぞれの可能な“値ごとの確率”を全部あらわす関数。
- **累積分布関数（CDF）**：「ここまでにうまくいく確率をぜんぶ足しあわせたもの」。
- **変数 $X$**：ここでは「初めて当たるまでの回数」という意味で使っている。

---

### 実生活で役立ちそうな場面

- おみくじで「3回までに大吉が出るチャンスは？」
- ゲームで「5回以内にレアアイテムがドロップする確率って？」
- 仕事で「10人営業電話して1件でも契約がとれる確率は？」

…こうした「これまでの合計“成功までの確率”や、“きっかけになる1回”の確率」を計算できる。

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import geom

# --- このコードの目的・背景 ---
# 幾何分布（geometric distribution）の確率質量関数（PMF）がどのような形になるかを、
# 「成功確率 p の値を変えながら視覚的に比較」するためのサンプルコードである。
# たとえば、「コイン投げ（p=0.5）」・「あまり当たらないくじ（p=0.1）」など
# 条件によって幾何分布のカーブがどう変化するのか？を知ることが目的。

# --- パラメータ説明 ---
# p_values: 成功確率pのリスト。pは「1回の試行で成功する確率」
# 例）本当に公平なコインなら p=0.5
#     20%の確率で始めて成功できるゲームなら p=0.2
#     レアアイテムが10%で手に入るなら p=0.1
p_values = [0.5, 0.2, 0.1]

# k_values: 試行回数kのリスト。kは「初めて成功するまで何回チャレンジしたか」
# 例）k=1: 1回でうまくいった, k=2: 2回目で初めて成功, ... k=15まで考える
k_values = np.arange(1, 16)  # 1から15までの試行回数

plt.figure(figsize=(12, 5))  # グラフのサイズを広めに

# --- 各成功確率pごとにPMFカーブを計算・表示する（forループで繰り返し） ---
for i, p in enumerate(p_values):
    # geom.pmf(k_values, p): 幾何分布のPMFをk=1～15について計算
    # それぞれ、「k回目で初めて成功する確率」を返している
    pmf_values = geom.pmf(k_values, p)
    
    # ラインチャートで可視化（o-で点線＆マーカー）
    plt.plot(k_values, pmf_values, 'o-', label=f'p = {p}')

# --- グラフの外観カスタマイズ ---
plt.title('幾何分布の確率質量関数 (PMF)')
plt.xlabel('初めて成功するまでの試行回数 k')
plt.ylabel('確率 P(X=k)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

# ---------------------------- コメントまとめ -----------------------------
# ・何をしている？→ 幾何分布でp=0.5, 0.2, 0.1 のとき、PMFの形がどうなるか描画
# ・なぜやるの？→ 成功確率が大きいほど「早めに成功」が多く、確率が小さいほど
#   「なかなか成功しないケースが増える（曲線が右に伸びる）」のを可視化できる
# ・得られる理解→ 実際にグラフで見ることで「何回目で成功しやすいか」の違いを直感で掴める
# ---------------------------- 変数例 -----------------------------
# p_values = [0.5, 0.2, 0.1]    # コイン(50%)、くじ20%、レアアイテム10%
# k_values = [1, 2, ..., 15]    # 最短1回め～長ければ15回めまで
# ----------------------------------------------------------------
```

## 3.4.3 期待値と分散（やさしい解説）

### 期待値（平均）とは？

「期待値（きたいち）」とは、「平均すると何回で成功できそうか？」を表す言葉である。  
たくさん試してみたときに、成功するまでにいつも何回くらいかかるか、その“だいたいの平均値”のこと。

幾何分布では、期待値はとてもシンプルに式で書ける。

$$
E[X] = \frac{1}{p}
$$

ここで
- $E[X]$ は期待値（平均試行回数）
- $p$ は「1回で成功する確率」

#### 例でイメージしよう
- コイン投げ（$p=0.5$）：平均 $1/0.5 = 2$ 回で表が出る
- サイコロで6（$p=1/6$）：平均 $1/(1/6) = 6$ 回で6が出る
- 10%で当たりが出るくじ（$p=0.1$）：平均 $1/0.1 = 10$ 回で当たりが出る

**イメージ:**  
「初めて当たるまでに、だいたい何回がんばれば良いか？」それが期待値である。

---

### 分散とは？

「分散（ぶんさん）」は、「どれくらい結果にバラつきがあるか」を表す数字である。

期待値が同じでも、実際には「すぐに当たる人」もいれば「全然当たらない人」もいる。そのバラつきの大きさを知りたいときに分散を使う。

幾何分布の分散はこうなる。

$$
Var[X] = \frac{1-p}{p^2}
$$

- $Var[X]$ は分散
- $p$ は1回で成功する確率

**ポイント:**  
$p$ が小さくなる（なかなか当たらない場合）は分散が大きくなり、何回かかるかの予測がむずかしくなる。

---

### 標準偏差（ひょうじゅんへんさ）

「標準偏差」は、分散を「元の単位（ここでは回数）」に戻すために使う数字。  
分散が「バラつきの平均の2乗」なら、標準偏差は「ふつうのバラつき」を直接数で表してくれる。

$$
\sigma = \sqrt{Var[X]} = \sqrt{\frac{1-p}{p^2}} = \frac{\sqrt{1-p}}{p}
$$

- $\sigma$（シグマと読む）が標準偏差

---

### 計算例で実感！

#### コイン投げ（$p=0.5$）
- 期待値：$E[X] = 1 / 0.5 = 2$
  - 平均で2回くらい投げて表が出る
- 分散：$Var[X] = (1-0.5) / (0.5^2) = 0.5 / 0.25 = 2$
- 標準偏差：$\sqrt{2} \approx 1.414$
  - 実際は2回の前後にふらつく

#### サイコロで6の目（$p=1/6$）
- 期待値：$E[X] = 1 / (1/6) = 6$
  - 平均で6回目くらいで6が出る
- 分散：$Var[X] = (1-1/6) / (1/6)^2 = (5/6) / (1/36) = 30$
  - バラつきはかなり大きい
- 標準偏差：$\sqrt{30} \approx 5.477$
  - 6回目より大きく外れる場合もけっこうある

---

### 用語のやさしい解説

- **期待値**：平均するとどれくらいか、のメインの値。数学では$E[X]$と書く。
- **分散**：バラつきの度合い。どれくらい「予想と違うことが起きるか」を数字で表す。
- **標準偏差**：「分散」の平方根。実際の「ずれ」をイメージしやすくするための値。

---

### 実生活で役立つ場面

- まとめて何回トライすれば1回は当たる？営業・くじ引き・機械の故障検査など、**平均で何回必要か**を知りたいときにすごく使える。
- 変動が大きいか小さいか（分散）、つまり「運に左右されやすいか」を知ることで、戦略を立てたり不安を和らげたりできる。

```python
# --- このコードの意図・背景 ---
# 幾何分布（geometric distribution）において「期待値（平均）」「分散」「標準偏差」を
# 成功確率pごとに計算・比較するためのプログラムである。
#
# ・幾何分布は「はじめて成功するまでのチャレンジ回数」のまわりに使う分布。
# ・成功確率pによって、「何回で当たる？」「バラつきはどれくらい？」が大きく変わる。
#
# このコードは p の値をいろいろ変えて、「平均やバラつきがどうなるか？」を
# 実際に数字で示すことが目的である。

import numpy as np  # np.sqrt を使うため

# --- 幾何分布の期待値、分散、標準偏差を計算する関数 ---
def geometric_stats(p):
    # 期待値（平均回数）
    mean = 1/p
    # 分散（バラつきの2乗）
    variance = (1-p)/(p**2)
    # 標準偏差（バラつきの代表的な数値、分散の平方根）
    std_dev = np.sqrt(variance)
    return mean, variance, std_dev

# --- 計算したい成功確率pと、その説明ラベルをセット ---
p_examples = [0.5, 0.2, 0.1, 1/6]
#     ↑ p=0.5: 公平コイン表
#       p=0.2: 20%で当たり
#       p=0.1: 10%で当たり
#       p=1/6: サイコロで6が出る
# ほか、p=0.01（1%で当たる超レア）なども考えられる
p_labels = ["0.5 (コイン)", "0.2 (20%の確率)", "0.1 (10%の確率)", "1/6 (サイコロの6)"]

# --- 結果をきれいに表の形式で出力 ---
print("幾何分布の期待値と分散")
print("-" * 50)
print(f"{'成功確率 p':<15} {'期待値 E[X]':<15} {'分散 Var[X]':<15} {'標準偏差 σ':<15}")
print("-" * 50)

for p, label in zip(p_examples, p_labels):
    mean, var, std = geometric_stats(p)
    # 小数2桁で美しく表示（<15で揃え）
    print(f"{label:<15} {mean:<15.2f} {var:<15.2f} {std:<15.2f}")

# --- コメントまとめ ---
# ・目的：いろんな「当たりやすさ（成功確率）」で、平均チャレンジ回数やブレの大きさが
#      どう違うか見やすく一覧化したい
# ・p(成功確率)が大きい　→　期待値（必要な回数）も小さい（たとえばコインは平均2回）
# ・pが小さい（当たりにくい）→　期待値も分散も標準偏差も大きくなる
#   （例: p=0.1なら「平均10回」だが、実際には20回以上かかることも珍しくない）
# ・現実では「営業成約」「くじ」「障害発生○回目」の予測や計画などにも役立つ
```

### 幾何分布の「期待値」「分散」「標準偏差」の読み解き方とσの意味

#### 1. 表の見方：ポイント解説

| 成功確率 p       | 期待値 E[X] | 分散 Var[X] | 標準偏差 σ |
|-----------------|-------------|--------------|-------------|
| 0.5 (コイン)    | 2.00        | 2.00         | 1.41        |
| 0.2 (20%の確率) | 5.00        | 20.00        | 4.47        |
| 0.1 (10%の確率) | 10.00       | 90.00        | 9.49        |
| 1/6 (サイコロの6)| 6.00        | 30.00        | 5.48        |


- **期待値E[X]**：「平均すると何回で初めて成功できるか？」の目安。
  - 例：サイコロの6（p=1/6）なら、平均して6回ふれば1回は6が出る。
- **分散Var[X]**：「毎回の結果が平均からどれだけブレるか」の2乗の平均。数値的イメージはつきにくいので、感覚的には標準偏差を見るとよい。
- **標準偏差σ**：「だいたいいつも、期待値からどれくらいズレることが多いか」の目安。
  - 例：サイコロの6なら、期待値6回に対し、標準偏差5.48回ぶんズレる場合が多い。
    - 期待値6回（＝平均すると6回で6が出る）
    - 標準偏差5.48回（＝だいたいの場合は6回から**プラスマイナス5.48回ぶんズレる** 、「6 - 5.48」回や「6 + 5.48」回ぐらいの範囲がわりと出てくる）
    - **実際に成功するまでにかかる回数Xは、期待値E[X]=6回の“まわり”に、標準偏差σ=5.48回ぶんの広がりで分布している**。   つまり、“ふつうは平均６回からだいたい5～6回ぶん前後にズレることが多い”というイメージになる。
    

---

#### 2. 標準偏差「σ」が大きいってどういうこと？

##### 疑問：  
> 期待値E[X]が6.00に対して、σが5.48もあります。これはどう捉えればいいのでしょう？

##### 回答：

- 標準偏差$\sigma$は、「実際にかかる回数が“期待値の前後でどのくらい広がるか”」を表す指標。
- 期待値（=平均）は「だいたいの中心値」だが、「毎回必ずこれにピタリとなる」わけではない。
- 今回のサイコロ例（期待値6、σ約5.5）の場合…
    - 実際には6回で6が出る場合もあれば、**それよりかなり早い/遅い**ことも多い、という意味。
    - ざっくり「6回プラスマイナス5~6回ぶんくらい大きくブレることがかなりある」イメージ。
- 幾何分布では **「運が悪いと失敗が続き、極端に時間がかかることも珍しくない」** ため、標準偏差が大きくなることが特徴。
- 特に成功確率が低い（レアな出来事）ほど、「なかなか当たらない人」「すぐ当たる人」の差が広がる。

---

#### 3. この結果をどう生かせる？

- **平均だけで考えると危険**  
  期待値6回、標準偏差5.5回なので、「毎回6回で確実に当たる」わけではない。  
  たとえばイベントや業務計画で「平均6回なんだから必ず6回やれば出る！」と穴がないと思いこむのは注意。
- **ばらつきに備える**  
  何回くらい余分に用意すればだいたい目標達成できるか、標準偏差を目安に「安全マージン」を考えるのが現実的戦略につながる。
- **まとめ**  
  幾何分布では「“平均”よりずっと早く or ずっと遅く」当たりが出ることもよくある。  
  標準偏差が大きいほど、「結果が大きくバラつく」ことを意味する。

---

#### 用語やポイントのさらなる解説

- **期待値E[X]** ：「平均」。長い目で回数を重ねたときの中心的な回数。
- **標準偏差σ** ：「ふつうは平均からどれくらいズレるか」の目安。σがE[X]と同じくらい大きい場合は、かなり幅が広い分布になる。
- **例**：「サイコロ6回目で6が出る」のは“平均”だが、1回で6が出たり、20回以上かかったりも珍しくない。

> **要するに**  
> 幾何分布では「何回で初めて成功するか」は思ったより“ブレが大きい”ので、「平均」だけではなく「ばらつき（標準偏差）」も必ずチェックしよう！

## 3.4.4 幾何分布のシミュレーション



実際に幾何分布に従う乱数を生成し、理論的な分布と比較してみよう。



### シミュレーションの方法



幾何分布のシミュレーションは、次のように行うことができる：

1. 成功確率 $p$ を設定する
2. 各試行で乱数を生成し、$p$ より小さければ「成功」、そうでなければ「失敗」とする
3. 初めて「成功」が出るまでの試行回数を記録する
4. これを多数回繰り返し、試行回数の分布を調べる

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

# --- 目的 ---
# 幾何分布（初めて成功するまでにかかる回数）の乱数シミュレーションと、その理論値との比較を行う。
# ・実際に試してみた場合の分布と、
# ・計算上（理論上）出る分布
# がどれだけ一致するかを直感的につかむためのコード

# --- パラメータ設定 ---
p = 0.2             # 成功確率 p（1回試行したときの成功の確率。例：20%で当たり）
n_simulations = 10000   # シミュレーションの回数（たくさん試すことで本当の確率に近づく）

# --- 方法1: numpyの built-in 幾何分布乱数を使う ---
# geometric_samples: 各シミュレーションで「何回目で初成功するか」をランダム生成
geometric_samples = np.random.geometric(p, n_simulations)
# 例: [3, 1, 7, 2, ...] などの値が入り、「最初の成功まで繰り返した回数」がずらっと並ぶ

# --- 方法2: 手作りシミュレーション（ループで一つずつ） ---
def simulate_geometric(p, n):
    """
    幾何分布を乱数シミュレーションで「手作り」する関数。
    1回ごとに a.「失敗なら繰り返す」b.「成功したらカウント終了」を繰り返す。
    """
    results = []
    for _ in range(n):
        trials = 1  # 最初の試行
        while np.random.random() > p:  # np.random.random()は0～1の乱数。失敗なら続ける
            trials += 1  # もう一度トライ
        results.append(trials)  # 初成功までにかかった回数を追加
    return results

# 手作り関数でシミュレーション。1000回分だけ（時間短縮のため）。
manual_samples = simulate_geometric(p, 1000)

# --- 理論的な幾何分布のPMF（確率質量関数）値を計算 ---
k_range = np.arange(1, 16)       # 1回目～15回目までシミュレート
pmf_theoretical = geom.pmf(k_range, p)   # 理論的に各回数で「初成功」の出る確率

# --- ヒストグラムと理論値グラフを重ねて描画 ---
plt.figure(figsize=(12, 6))

# シミュレーション結果（numpy乱数利用）のヒストグラム（確率が分かりやすいよう density=True）
plt.hist(
    geometric_samples,             # サンプルデータ
    bins=range(1, 16),             # 1,2,3,...15回目まで区切って表示
    density=True,                  # 各binの高さ=確率になるよう正規化
    alpha=0.6,                     # 少し透明
    label='シミュレーション結果'    # 凡例
)

# 理論上のPMFと比較の赤い点・折れ線グラフ
plt.plot(
    k_range, pmf_theoretical, 
    'ro-', 
    label='理論的PMF'
)

# 理論的期待値（平均の試行回数）を縦線で表示（目安として）
mean_theoretical = 1/p
plt.axvline(
    mean_theoretical, 
    color='g', linestyle='--', 
    label=f'理論的期待値 = {mean_theoretical}'
)

# グラフの補助情報
plt.title(f'幾何分布のシミュレーション (p = {p}, n = {n_simulations})')
plt.xlabel('初めて成功するまでの試行回数 k')
plt.ylabel('確率 P(X=k)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

# --- 理論値 vs. 実験値の直接比較 ---
mean_empirical = np.mean(geometric_samples)         # シミュレーションの平均値
var_empirical = np.var(geometric_samples)           # シミュレーションの分散
mean_theoretical = 1/p                              # 理論上の期待値
var_theoretical = (1-p)/(p**2)                      # 理論上の分散

print(f"成功確率: p = {p}")
print("\n理論値と実験値の比較:")
print(f"期待値: 理論値 = {mean_theoretical:.4f}, 実験値 = {mean_empirical:.4f}")
print(f"分散: 理論値 = {var_theoretical:.4f}, 実験値 = {var_empirical:.4f}")

# --- コメントまとめ ---
# ・乱数シミュレーションと理論値がどれくらい近いか、目で見て・数字で比べて、体感できる
# ・pをいろいろ変えて試すと、「当たりにくいほど分布が右に伸びて、“なかなか当たらない人”も増える」こともわかる
# ・手作りシミュレーション（manual_samples）は学習用。numpyのgeometric関数は実用向け。
```

## 3.4.5 幾何分布の応用例

幾何分布は、「初めて成功するまでの試行回数」を表す確率分布として、いろいろな分野で役立っている。

---

### 品質管理

- **例の解説**：たとえば工場で同じ商品をどんどん作るとき、「何個目で初めて不良品（こわれたり傷が付いていたりする商品）が出てくるかな？」を予想する。  
- **簡単なイメージ**：1つ1つ見ていって、「ようやくハズレ品が見つかるまで、何個作ることになる？」  
- **数字の例**：1個ごとに5%の確率で不良品なら、平均で20個目に1つ見つかる。

---

### 信頼性工学

- **例の解説**：家電や自動車のような機械を使うとき、「何時間動かせば初めてこわれるか？」を平均で予想する。  
- **簡単なイメージ**：ゲーム機が故障するまで、どのくらい遊べる？  
- **数字の例**：1時間ごとに1%の確率で故障すると、平均100時間使うとどれかが故障。

---

### マーケティング

- **例の解説**：「何人のお客さんに説明すれば、やっと1人商品を買ってくれるかな？」の目安を知りたいときに使う。  
- **簡単なイメージ**：「営業電話で何回に1回契約がとれるんだろう？」  
- **数字の例**：1回話すごとに20%の確率で買ってもらえる場合、平均で5回声をかけて1人買ってくれる。

---

### ゲーム理論

- **例の解説**：ガチャやくじ、サイコロゲームなど、「当たりが出るまでに何回やることになる？」を数式で予想したいときに使われる。  
- **簡単なイメージ**：ガチャでレアアイテムが来るまで何回引けばいい？  
- **数字の例**：当たりが10%なら、平均10回に1回レアが出る。

---

### 待ち行列理論

- **例の解説**：銀行や役所などで、空いている窓口が来るまで「何人目で案内されるかな？」とか、「次に呼ばれるまでどれくらい待つ？」を予想するときに使う。  
- **簡単なイメージ**：ずっと行列にいて、いつサービスを受けられるか？  
- **数字の例**：1人につき10%の確率ですぐ窓口が空く場合、平均10人待ったら順番が回ってくる。

---

### 医学統計

- **例の解説**：薬が効きはじめるまで何日かかるか、治療を何度試したら効果が出るか――という「最初にうまくいくまで」を予想。  
- **簡単なイメージ**：治療を続けて、どれくらいで効果が現れる？  
- **数字の例**：1日ごとに10%の確率で効果が出るなら、平均10日で効く。

---

### 情報理論

- **例の解説**：コンピュータがデータをやり取りするとき、ミス（エラー）が最初に起こるまでどのくらい「送信」できるか、を数式で扱いたいとき。
- **簡単なイメージ**：データ通信でミスなく送れる平均回数や、エラー検出のタイミングを予想する。
- **数字の例**：1回のデータ送信で1%の確率でエラーが出るなら、平均100回送って1回はエラーになる。

---

> どの例も「ある確率で失敗や成功が起きる」という“繰り返しの中で、初めて当たりを引くまでの平均的な回数”を知りたいときに幾何分布が使われる。

## 3.4.6 まとめ



幾何分布は、「初めて成功するまでの試行回数」を表す離散確率分布である。以下の特徴を持つ：



- パラメータは1つだけ：成功確率 $p$
- 確率質量関数：$P(X = k) = (1-p)^{k-1} \cdot p$
- 期待値：$E[X] = \frac{1}{p}$
- 分散：$Var[X] = \frac{1-p}{p^2}$



幾何分布は、独立試行の繰り返しにおいて「初めての成功」を待つ状況を表現するのに適している。例えば、コインを投げて初めて表が出るまでの回数、くじ引きで当たりが出るまでの回数などである。



成功確率 $p$ が小さいほど、平均試行回数は大きくなり、また分散も大きくなる。つまり、成功確率が低いほど、成功するまでに多くの試行が必要になり、また結果のばらつきも大きくなる。



幾何分布は、品質管理、信頼性工学、マーケティング、ゲーム理論など、様々な分野で応用されている。

## 3.4.7 ちなみに 〜幾何分布はなぜ「幾何」なの？〜

「幾何分布（きかぶんぷ）」という名前の「幾何（geometric）」は、  
**「幾何級数（きかきゅうすう）」**（英語では geometric series）という数学の考え方に由来している。

---

### 幾何級数って何？

- 幾何級数とは、「ある数を何度もかけていく和」のこと。
- 例えば $a + ar + ar^2 + ar^3 + \ldots$ のように、
  毎回「前の項」に同じ数$r$をかけていく形で増えていく数列を指す。

### 幾何分布との関係

- 幾何分布の確率を計算するときに $(1-p)^{k-1} \cdot p$ の形があらわれる。
- たとえば「1回で失敗」→「2回で失敗」→・・・→「k-1回失敗＋k回目で成功」というふうに、
  「失敗を重ねるたび、確率が何度も掛け算でどんどん小さくなっていく」。
- この「掛けていく」「同じ率で増減する」という部分が「幾何（geometric）」と同じなのだ。

### まとめ

つまり**幾何分布の名前の理由は、「確率が幾何級数のように次々とかけ算で減っていく仕組み」から来ている**。  
分布自体に「図形（geometry）」の意味があるわけではないので注意しよう。
